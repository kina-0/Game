```python
import nbformat
import pygame
import random
import math
pygame.init()
screen=pygame.display.set_mode((800,600))#800=width and 600=height
#adding a background
background=pygame.image.load(r'C:\Users\win 10\Downloads\abstract-space-background-with-nebula-stars.jpg')
#changing color, title etc
pygame.display.set_caption("space invaders")
#player
playerImg=pygame.image.load(r'C:\Users\win 10\Downloads\spaceship.png')
#now we gonna give coordinates that where we want to position our player in the game 
playerX=370
playerY=480
player_Xchange=0

#enemy
enemyImg=[]
enemyX=[]
enemyY=[]
enemy_Xchange=[]
enemy_Ychange=[]
num_of_enemy=6
for i in range(num_of_enemy):
    
    enemyImg.append(pygame.image.load(r'C:\Users\win 10\Downloads\final-boss.png'))
#now we gonna give coordinates that where we want to position our player in the game 
    enemyX.append(random.randint(0,735))
    enemyY.append(random.randint(50,150))
    enemy_Xchange.append(4)
    enemy_Ychange.append(40)

#bullet
#ready= you cant see the bullet on the screen
#fire= bullet is currently moving
#these are the terms that we used to differentiate that if the bulleyts is in at rest or at state of fire
bulletImg=pygame.image.load(r'C:\Users\win 10\Downloads\bullet.png')
#now we gonna give coordinates that where we want to position our player in the game 
bulletX=0
bulletY=480 #bcz space ship is at 480 , nd we need to shoot the bullet from the spaceship
bullet_Xchange=0
bullet_Ychange=10
bullet_state="ready"
#score
score_value=0
font=pygame.font.Font('freesansbold.ttf',32)#.ttf is a extension of the file

textX=10
textY=10
#game over text
over_font=pygame.font.Font('freesansbold.ttf',32)#.ttf is a extension of the file


def show_score(x,y):
    score=font.render("Score: "+str(score_value),True,(255,255,255))
    screen.blit(score,(x,y))
    
def game_over_text():
    game_over_text=font.render("GAME OVER", True,(255,255,255))
    screen.blit(game_over_text,(300,250))
    
    

def player(x,y):
    #blit basically means to draw, we are drawimg image on our game surface
    #after uploading a image , we also need to draw it
    screen.blit(playerImg,(x,y))
    
    
def enemy(x,y,i):
    #blit basically means to draw, we are drawimg image on our game surface
    #after uploading a image , we also need to draw it
    screen.blit(enemyImg[i],(x,y))
    
    
def fire(x,y):#we want to fire the bullet whenever the spacebar is pressed
    global bullet_state
    bullet_state="fire"
    screen.blit(bulletImg,(int(x+16),int(y+10)))#we have add 10, 16, bcz we want bullet to be at centre of the spaceship

def collision(enemyX,enemyY,bulletX,bulletY):
    distance=math.sqrt((math.pow(enemyX-bulletX,2))+ (math.pow(enemyY-bulletY,2)))
    if distance<27:
        return True
    else:
        return False
    

    
c=True
while c==True: 
    #RGB-all colours are ,ade up of these 3 colours
    #so we have to give value in the bracket acc 
    screen.fill((0,0,0))
    #background
    screen.blit(background,(0,0))
    
    #screen.blit(playerImg,(playerX,playerY))
    for event in pygame.event.get():
        #this contains almost every event in the game 
        #like pressing any button or moving mouse
        if event.type == pygame.QUIT:
            c=False
        #if keystroke is pressed , check whether it is right or left
        if event.type==pygame.KEYDOWN:
            #print("key stroke is pressed")
            if event.key==pygame.K_LEFT:
                player_Xchange=-0.9
            if event.key==pygame.K_RIGHT:
                player_Xchange=+0.9
            if event.key==pygame.K_SPACE:
                bulletX=playerX
                fire(bulletX,bulletY)
                if bullet_state=="ready":
                    fire(playerX,bulletY)
                #playerX is the current position of bullet in x-axis 
                #bullet Y is position on y axis
                #this is the reason that bullet is moving along with the spaceship
                
                
                
                
        if event.type==pygame.KEYUP:
            if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                player_Xchange=0
                
                
                

            
                
    
    #we need to call player function here in the while loop bcz we want player in our ecery screem and want it to be in its updated position.
    #we will put player func after the screen. fill bcz first screen is made then player is made
    #if w eput player before screen.fill, then image would be drawn under the screen
    playerX=playerX+player_Xchange#checking of boundaries of space ship
    if playerX<=0:
        playerX=0
    elif playerX>=736:
        playerX=736
     #movement of enemy
    for i in range(num_of_enemy):
        #game over
        if enemyY[i]>300:
            for j in range(num_of_enemy):
                enemyY[j] = 2000
            game_over_text()
            break
            
            
                
        enemyX[i]=enemyX[i]+enemy_Xchange[i]
        if enemyX[i]<=0:
            enemy_Xchange[i]=1
            enemyY[i]+=enemy_Ychange[i]
        elif enemyX[i]>=736:
            enemy_Xchange[i]=-1
            enemyY[i]+=enemy_Ychange[i]
        collision_1=collision(enemyX[i],enemyY[i],bulletX,bulletY)
        if collision_1:
            bulletY=480
            bullet_stste="ready"
            score_value +=1
            enemyX[i]=random.randint(0,735)
            enemyY[i]=random.randint(50,150)
        
        enemy(enemyX[i],enemyY[i],i)
    
        
    #bullet movement
    #bullet appears only when we press space bar
    #we dont want our bullet to disappear very quickly
    #we want it to stat on the screen for a while
    #by default state remains=ready
    if bulletY<=0:
        bulletY=350
        bullet_state="ready"
    if bullet_state=="fire":
            #fire(playerX,bulletY)
            fire(bulletX,bulletY)
            bulletY-=bullet_Ychange#changing the position of the bullet, that is moving the bullet in upward direction ie.-y axis.
            
    player(playerX, playerY)
    show_score(textX,textY)
    pygame.display.update()
        
```
